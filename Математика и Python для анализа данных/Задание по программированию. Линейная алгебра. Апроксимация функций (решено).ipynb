```python
import math
from matplotlib import pylab as plt
import numpy as np
from scipy import linalg
import scipy as sp

def f(x):
    return np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2)

x = np.arange(1, 15, 0.1)
a = ([1., 1.],[1., 15.])
b = ([f(1.),f(15.)])
c1 = sp.linalg.solve(a,b)

a = ([1.,1.,1.],[1.,8.,8.**2],[1,15.,15**2])
b = ([f(1.),f(8.),f(15.)])
c2 = sp.linalg.solve(a,b)

a = ([1.,1.,1.,1.],[1.,4.,4.**2,4**3],[1.,10.,10.**2,10**3],[1.,15.,15.**2,15.**3])
b = ([f(1.),f(4.),f(10.), f(15.)])
c3 = sp.linalg.solve(a,b)
print (c3)

a = ([1.,1.,1.,1.,1.],[1.,4.,4.**2,4**3,4**4],[1.,7.,7.**2,7**3,7**4],[1.,10.,10.**2,10**3,10**4],[1.,15.,15.**2,15.**3,15**4])
b = ([f(1.),f(4.),f(7.), f(10.), f(15.)])
c4 = sp.linalg.solve(a,b)


y = np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2)
solve1 = c1[0]+c1[1]*x
solve2 = c2[0]+c2[1]*x+c2[2]*x**2
solve3 = c3[0]+c3[1]*x+c3[2]*x**2+c3[3]*x**3
solve4 = c4[0]+c4[1]*x+c4[2]*x**2+c4[3]*x**3+c4[4]*x**4

plt.plot(x, y)
plt.plot(x, solve1)
plt.plot(x, solve2)
plt.plot(x, solve3)
plt.plot(x, solve4)
plt.show()
```
